## Data Cleaning

```python
import pandas as pd
```

```python
data = pd.read_csv('C:\\Users\\user\\Documents\\datasets\\tweet.csv')
```

```python
data['Replies'] = data['Replies'].str[:-1]
data['Retweets'] = data['Retweets'].str[:-1]
data['Likes'] = data['Likes'].str[:-1]
```

```python
data
```

```python
data['Replies'] = (pd.to_numeric(data['Replies'].str.replace(',',''), errors='coerce'))
data['Retweets'] = (pd.to_numeric(data['Retweets'].str.replace(',',''), errors='coerce'))
data['Likes'] = pd.to_numeric(data['Likes'])
```

```python
data
```

```python
data['Replies'] = (data['Replies'])*1000
data['Retweets'] = (data['Retweets'])*1000
data['Likes'] = (data['Likes'])*1000
```

```python
data
```

```python
data['Timestamps'] = pd.to_datetime(data['Timestamps'])
```

```python
data
```

```python
data.to_csv('C:\\Users\\user\\Documents\\datasets\\cr7_tweet.csv',index = False)
```
